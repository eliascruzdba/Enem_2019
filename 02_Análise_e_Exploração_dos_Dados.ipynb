# Substituir Notas faltantes pela Média

Temos duas opções, realizar o drop dos valores "NaN" ou substituir os valores faltantes pela média o que evita perda de informações importantes.

Visualização dos atributos do dataframe que são diferentes de Object

```python
# Campos nulos
Nulos = base.isnull()
plt.figure( figsize=(16,5 ) )
plt.title('Analise campos nulos')
sns.heatmap( Nulos, cbar=False );
```

Utilizamos o select_dtypes para obter somente os dados numéricos através da opção "exclude='object'".

```python
base.select_dtypes(exclude='object')
```

```python
#aplicando a média para o target de dados faltantes
base['NU_NOTA_CN'].fillna((base['NU_NOTA_CN'].mean()), inplace=True)
base['NU_NOTA_CH'].fillna((base['NU_NOTA_CH'].mean()), inplace=True)
base['NU_NOTA_LC'].fillna((base['NU_NOTA_LC'].mean()), inplace=True)
base['NU_NOTA_MT'].fillna((base['NU_NOTA_MT'].mean()), inplace=True)
base['NU_NOTA_REDACAO'].fillna((base['NU_NOTA_REDACAO'].mean()), inplace=True)
```

```python
#incluindo o min para o dado faltante.
base['TP_DEPENDENCIA_ADM_ESC'].fillna((base['TP_DEPENDENCIA_ADM_ESC'].min()), inplace=True)
```

```python
# Campos nulos
Nulos = base.isnull()
plt.figure( figsize=(16,5 ) )
plt.title('Analise campos nulos')
sns.heatmap( Nulos, cbar=False );
```

Removendo valores = 0, indicando que o candidado zerou a prova ou não compareceu. Esse tipo de tratamento evitara alguns outliers o que pode nos ajudar quando executarmos os modelos de Machine Learning.

# SEPARANDO BASES TREINAMENTO E TESTE

Um conceito importante em machine learning é a divisão dos dados em dois grupos:

*   dados de treinamento do modelo
*   dados de checagem do modelo

É muito importante fazer isso antes de iniciar a construção do modelo, pois isso evita que o resultado utilizado para treinamento do modelo seja o mesmo de checagem, gerando-se assim uma métrica errada de qualidade do modelo, isto é um modelo pode parecer muito bom para dados de treinamento, porém pode ser muito ruim para dados fora da amostra de treinamento.

Para isso o python possui uma biblioteca que facilita bastante a vida do usuário: sckit-learn.

```python
from pandas.core.common import random_state
# Splitting dataset into training and testing
from sklearn.model_selection import train_test_split
seed = 4321
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=seed)
```

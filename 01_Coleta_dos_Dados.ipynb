### Carregando uma porcentagem da base de dados </br>

Realizando uma primeira visualização temos uma quantidade de dados enorme. Visando economizar o poder computacional vamos utilizar apenas uma porcentagem dos dados obtidos.

Link: https://drive.google.com/file/d/1hg1PKOht8TLnaiUHNFBAgQQqUBU7To-i/view?usp=sharing

O total de linhas da base é 5095271, resolvemos diminuir para que o processamento não seja impactado. </br>
Total de linhas para Estudo: 76428 e 136 colunas.

```python
#arquivo
arquivo = '../content/drive/MyDrive/Colab Notebooks/Dataframe/df_ENEM_2019.csv' #<<<<<<-----Arquivo compartilhado via Drive

# define o padrao de aleatoriedade
seed = 31123 
np.random.seed(seed)

#conta quantas linhas tem o data set
total_linhas = sum(1 for l in open(arquivo,encoding="ISO-8859-1") )
print(f'Total de Linhas: {total_linhas}')

#calcula quanto é 0.5% dos dados 
#p = 0.0017
p = 0.015
tamanho = int(total_linhas*p)
print(f'Total de Linhas para Estudo: {tamanho}')

# lista de numeros aleatorios
index = random.sample(range(1, total_linhas), total_linhas - tamanho)

# selecions linhas da base de dados com base na lista de numero aleatórios
df = pd.read_csv(arquivo,encoding="ISO-8859-1",delimiter=";",skiprows=index)#, usecols=colunas)

df.shape
```

```python
base_media = base.filter(items=['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
base_media
```

agregar informações por Município

```python
Notas_Enem_Municipio = base_media.groupby(['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA'], as_index=False).mean()
```

Podemos perceber que possuimos dados NaN em relação as notas. Vamos aplicar o "fillna" como tratativa dessas informações.

```python
Notas_Enem_Municipio.isna().sum()
```

Realizando preenchimento de dados Faltantes com a Média

Utilizando dessa forma para não ter que deletar dados.

```python
#aplicando a média para os dados faltantes
Notas_Enem_Municipio['NU_NOTA_CN'].fillna((Notas_Enem_Municipio['NU_NOTA_CN'].mean()), inplace=True)
Notas_Enem_Municipio['NU_NOTA_CH'].fillna((Notas_Enem_Municipio['NU_NOTA_CH'].mean()), inplace=True)
Notas_Enem_Municipio['NU_NOTA_LC'].fillna((Notas_Enem_Municipio['NU_NOTA_LC'].mean()), inplace=True)
Notas_Enem_Municipio['NU_NOTA_MT'].fillna((Notas_Enem_Municipio['NU_NOTA_MT'].mean()), inplace=True)
Notas_Enem_Municipio['NU_NOTA_REDACAO'].fillna((Notas_Enem_Municipio['NU_NOTA_REDACAO'].mean()), inplace=True)
```

```python
Notas_Enem_Municipio.isna().sum()
```

```python
Notas_Enem_Municipio
```

Faremos a junção das bases utilizando a biblioteca pandas "pd.merge".

```python
Media_Enem_Municipio_Idhm = pd.merge(Notas_Enem_Municipio, Atlas_Humano, how='left', left_on=['CO_MUNICIPIO_RESIDENCIA'], right_on=['Codmun7'] )
```

```python
Media_Enem_Municipio_Idhm.dtypes
```

Como o Python utiliza ponto ao invés de vírgula vamos utilizar um tratamento para as colunas IDHM estão como float64, aplicando o "replace" passando o valor de ',' para '.'. </br>

Exemplo:</br>numero = '0,99'</br>
>str(numero).replace(",",".") 
>>numero = '0.99'

```python
#Tratamento realizando um replace de ,="Vírgula" por .="ponto".
Media_Enem_Municipio_Idhm['IDHM'] = Media_Enem_Municipio_Idhm['IDHM'].apply(lambda x: float(str(x).replace(',','.')))
Media_Enem_Municipio_Idhm['IDHM_E'] = Media_Enem_Municipio_Idhm['IDHM_E'].apply(lambda x: float(str(x).replace(',','.')))
Media_Enem_Municipio_Idhm['IDHM_L'] = Media_Enem_Municipio_Idhm['IDHM_L'].apply(lambda x: float(str(x).replace(',','.')))
Media_Enem_Municipio_Idhm['IDHM_R'] = Media_Enem_Municipio_Idhm['IDHM_R'].apply(lambda x: float(str(x).replace(',','.')))
```
